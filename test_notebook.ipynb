```python
import sympy
import numpy as np
from matplotlib import pyplot as plt

normalize=False

pi = sympy.pi.evalf(400)

digits = (d for d in str(pi)[2:])

freqs = np.zeros(10, dtype='i4')

for d in digits:

    freqs[int(d)] += 1

if normalize:

    freqs = freqs/freqs.sum()
    

ax = plt.plot(freqs,'bo-')

plt.title('Single digit counts in pi')

plt.xlabel('Digit')

plt.ylabel('Count')
```
